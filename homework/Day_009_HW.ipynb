```python
# -*- coding: utf-8 -*-
"""
Created on Sat Sep  7 13:55:06 2019

@author: kurage
"""
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

#load data
dir_data = 'C:/Users/kurage/Desktop/3rd100/Part01'
f_app = os.path.join(dir_data,'application_train.csv')
app_train = pd.read_csv(f_app)

#篩選數值型的欄位
a =app_train.dtypes.value_counts()

dtype_select = [np.dtype('int64'),np.dtype('float64')]
numeric_columns = list(app_train.columns[list(app_train.dtypes.isin(dtype_select))])

# 再把只有 2 值 (通常是 0,1) 的欄位去掉
numeric_columns = list(app_train[numeric_columns].columns[list(app_train[numeric_columns].apply(lambda x:len(x.unique())!=2 ))])
print("Numbers of remain columns" , len(numeric_columns))

"""
#檢視欄位的數值範圍
num_bins = 10
for col in numeric_columns:
    x = np.array(app_train[col],float)
    plt.hist(x,num_bins,edgecolor='k')
    plt.title(col)
    plt.xlabel("index")
    plt.ylabel("numbers")
    plt.show()
"""
"""
依照出來圖表,有問題的是
而會造成這些問題通常為系統錯誤或人為因素居多
# AMT_INCOME_TOTAL
# REGION_POPULATION_RELATIVE
# OBS_60_CNT_SOCIAL_CIRCLE
---
# OBS_30_SOCIAL_CIRCLE
# DEF_30CNT_SOCIAL_CIRCLE
# AMT_REQ_CREDIT_BUREAU_QRT
"""
# 最大值離平均與中位數很遠
print(app_train['AMT_INCOME_TOTAL'].describe())

# 繪製 Empirical Cumulative Density Plot (ECDF)

cdf=app_train['OBS_60_CNT_SOCIAL_CIRCLE']

plt.plot(list(cdf.index), cdf/cdf.max())
plt.xlabel('Value')
plt.ylabel('ECDF')
plt.xlim([cdf.index.min(),cdf.index.max()*1.05])#限制顯示圖片的範圍
plt.ylim([-0.05,1.05])#限制顯示圖片的範圍
plt.show()

#改變y軸的Scale,使的可正常顯示ECDF
plt.plot(np.log(list(cdf.index)),cdf/cdf.max())
plt.xlabel('Value(log-scale)')
plt.ylabel('ECDF')
plt.ylim([-0.05,1.05])
plt.show()

```
